## Feature selection

I chose the root temperature (pm) as the target variable. 
From the data description: _Especially rotor temperature and torque are not reliably and economically measurable in a commercial vehicle. Being able to have strong estimators for the rotor temperature helps the automotive industry to manufacture motors with less material and enables control strategies to utilize the motor to its maximum capability._

```python
target = 'pm'

df.corr()[target].abs().sort_values(ascending=False)
```

As shown above, stator_tooth, stator_winding, stator_yoke, and ambient are the most important features for predicting the rotor temperature. They are all measures of the temperature of different parts of the stator, which is the stationary part of the motor. The stator is responsible for creating a rotating magnetic field that interacts with the rotor to generate torque. As the stator heats up, it can affect the performance and efficiency of the motor.

The ambient temperature, which is the temperature of the surrounding environment, is also an important predictor of the rotor temperature. Changes in the ambient temperature can impact the heat dissipation of the motor and affect its thermal behavior.

## Model training

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from LinearRegressionCustom import LinearRegression as LinearRegressionCustom
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
def print_score(y_test, y_pred, score):
    print('MSE:', mean_squared_error(y_pred, y_test))
    print('MAE:', mean_absolute_error(y_pred, y_test))
    print('Score:', score)
```

```python
y = df[target]
```

### On the given data

```python
X = df.drop(target, axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
score = model.score(X_test, y_test)

print_score(y_test, y_pred, score)
```

```python
# Otherwise, I get an error
# ValueError: all the input arrays must have same number of dimensions, but the array at index 0 has 2 dimension(s) and the array at index 1 has 1 dimension(s)
y_train = y_train.values.reshape(-1, 1)

model = LinearRegressionCustom()
model.gradientDescent(X_train, y_train)
y_pred = model.predict(X_test)
print(y_pred)

score = r2_score(y_test, y_pred)

print_score(y_test, y_pred, score)
```

### On the train subset

```python
# Columns with an absolute correlation between 0.5 and 0.8 with the target column
features = ['stator_winding', 'stator_yoke', 'ambient']
X = df[features]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
score = model.score(X_test, y_test)

print_score(y_test, y_pred, score)
```

```python
# Otherwise, I get an error
# ValueError: all the input arrays must have same number of dimensions, but the array at index 0 has 2 dimension(s) and the array at index 1 has 1 dimension(s)
y_train = y_train.values.reshape(-1, 1)

model = LinearRegressionCustom()
model.gradientDescent(X_train, y_train)
y_pred = model.predict(X_test)
score = r2_score(y_test, y_pred)

print_score(y_test, y_pred, score)
```
